```python
# Cell 1: Setup
import requests
import pandas as pd
import re
import time
```

```python
# Cell 2: Mastodon API Data Collection

# 1. Configuration
INSTANCE = "mastodon.social" 
HASHTAG = "IranWar"             # Topic hashtag
LIMIT_REQUIRED = 2000
collected_posts = []
max_id = None                # Used for pagination (getting the next page)

print(f"Collecting posts from {INSTANCE} for #{HASHTAG}...")

# 2. Collection Loop
while len(collected_posts) < LIMIT_REQUIRED:
    # Construct the API URL
    url = f"https://{INSTANCE}/api/v1/timelines/tag/{HASHTAG}"
    params = {'limit': 50}
    if max_id:
        params['max_id'] = max_id
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        if not data:
            break
            
        for post in data:
            # Clean HTML tags from the post content
            content = re.sub('<[^<]+?>', '', post['content'])
            
            # Filter for posts that are long enough to have sentiment
            if len(content) > 30:
                collected_posts.append({
                    "created_at": post['created_at'],
                    "username": post['account']['username'],
                    "sentiment_text": content.replace("\n", " ").strip(),
                    "url": post['url']
                })
        
        # Set max_id to the last post's ID to get older posts in the next loop
        max_id = data[-1]['id']
        print(f"Gathered {len(collected_posts)} records so far...")
        
        # Ethical pause to avoid hitting rate limits
        time.sleep(1) 
    else:
        print(f"Error: {response.status_code}")
        break

# 3. Create DataFrame
df = pd.DataFrame(collected_posts).head(LIMIT_REQUIRED)
print(f"\nFinal count: {len(df)} records collected.")
```

```python
# Cell 3: Export Deliverable
if not df.empty:
    file_name = "Iran_War_Sentiment.csv"
    df.to_csv(file_name, index=False, encoding='utf-8')
    print(f"Success! '{file_name}' created.")
    
    # Display the data
    display(df[['username', 'sentiment_text']].head(10))
else:
    print("No data collected. Try a different instance or hashtag.")
```
